```python
data.dropna('index', 'any', inplace=True)
print(len(data))
```

```python
dateDict = {}
for y in range(len(sp)):
    x = sp['Date'][y]
    dateDict[x] = datetime.date(int(x[0:4]), int(x[5:7]), int(x[8:10]))
    
for fld in ['filedate', 'prevfiledate', 'nextfiledate']:
    for y in range(len(data)):
        try:
            x = data[fld][y]
            dateDict[x] = datetime.date(int(x[0:4]), int(x[5:7]), int(x[8:10]))
        except Exception:
            pass
        
len(dateDict)
```

```python
dateDict['2017-04-14'] = datetime.date(2017,4,14)
```

```python
# want: stock + market return since previous filedate, stock + market return until next filedate,
# did stock beat market for both periods (latter is target label)

def getAvailDate(ticker, maxDate):
    avail = [dateDict[x[1]]
                 for x in stockdata[ticker]['datatable']['data']
                 if dateDict[x[1]] <= maxDate]
    if avail:
        return max(avail)
    else:
        return None
    
savailcList = []
savailpList = []
savailnList = []
droprows = []
cache = {}
for x in range(len(data)):
#for x in range(1000):
    ticker = data['ticker'][x]
    cres = cache.get((ticker, data['filedateobj'][x]), None)
    if cres:
        savailc = cres
    else:
        savailc = getAvailDate(ticker, data['filedateobj'][x])
        cache[(ticker, data['filedateobj'][x])] = savailc
    cres = cache.get((ticker, data['prevfiledateobj'][x]), None)
    if cres:
        savailp = cres
    else:
        savailp = getAvailDate(ticker, data['prevfiledateobj'][x])
        cache[(ticker, data['prevfiledateobj'][x])] = savailp
    cres = cache.get((ticker, data['nextfiledateobj'][x]), None)
    if cres:
        savailn = cres
    else:
        savailn = getAvailDate(ticker, data['nextfiledateobj'][x])
        cache[(ticker, data['nextfiledateobj'][x])] = savailn
    if not (savailc and savailp and savailn):
        droprows.append(x)
        continue
    savailcList.append(savailc)
    savailpList.append(savailp)
    savailnList.append(savailn)
    
```

```python
print(len(data))
data.drop(data.index[droprows], inplace=True)
droprows = []
print(len(data))

data = data.reset_index(drop=True)
```

```python
stockpriceDict = {}
for t in stockdata.keys():
    for d in stockdata[t]['datatable']['data']:
        stockpriceDict[(t, d[1])] = d[5]
len(stockpriceDict)
```

```python
marketpriceDict = {}
for x in range(len(sp)):
    marketpriceDict[sp['Date'][x]] = sp['Adj Close'][x]
len(marketpriceDict)
```

```python
testdata = data[0:100]
spriceCurrList = []
spricePrevList = []
spriceNextList = []
mpriceCurrList = []
mpricePrevList = []
mpriceNextList = []
for x in range(len(data)):
    t = data['ticker'][x]
    spriceCurrList.append(stockpriceDict[(t, str(savailcList[x]))])
    spricePrevList.append(stockpriceDict[(t, str(savailpList[x]))])
    spriceNextList.append(stockpriceDict[(t, str(savailnList[x]))])
    mpriceCurrList.append(marketpriceDict[str(savailcList[x])])
    mpricePrevList.append(marketpriceDict[str(savailpList[x])])
    mpriceNextList.append(marketpriceDict[str(savailnList[x])])
```

```python
data['stockprice0'] = pandas.Series(spriceCurrList)
data['stockprice1'] = pandas.Series(spricePrevList)
data['stockpricefuture'] = pandas.Series(spriceNextList)
data['marketprice0'] = pandas.Series(mpriceCurrList)
data['marketprice1'] = pandas.Series(mpricePrevList)
data['marketpricefuture'] = pandas.Series(mpriceNextList)
data.head()
```

```python
import numpy as np
data['stockreturn0'] = np.log(data['stockprice0'] / data['stockprice1'])
data['stockreturnfuture'] = np.log(data['stockpricefuture'] / data['stockprice0'])
data['marketreturn0'] = np.log(data['marketprice0'] / data['marketprice1'])
data['marketreturnfuture'] = np.log(data['marketpricefuture'] / data['marketprice0'])
data['beatmarket0'] = data['stockreturn0'] > data['marketreturn0']
data['beatmarketfuture'] = data['stockreturnfuture'] > data['marketreturnfuture']
data.head()
```

```python
with open("D:/w266project/singledataset/cleanedandenhanced.csv", "w") as pfile:
    data.to_csv(pfile, index=False)
```
